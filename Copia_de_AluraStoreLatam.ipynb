### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt


url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

```

```python
tienda.head()
```

#1. Análisis de facturación



```python
def facturacion_total(tienda):
  return  float(tienda["Precio"].sum())

fact_tienda1 = facturacion_total(tienda)
fact_tienda2 = facturacion_total(tienda2)
fact_tienda3 = facturacion_total(tienda3)
fact_tienda4 = facturacion_total(tienda4)

facturacion_tiendas = pd.DataFrame({'Tiendas':['Tienda1','Tienda2','Tienda3','Tienda4'],
                                    'Facturacion':[fact_tienda1,fact_tienda2,fact_tienda3,fact_tienda4]})
facturacion_tiendas.plot(kind="bar", x='Tiendas', y='Facturacion', color= 'blue', title='Facturacion total de tiendas')
plt.xlabel("Tienda")
plt.ylabel("Facturación total")
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
def ventas_categorias(tienda):
  return tienda.groupby("Categoría del Producto")["Precio"].sum()

ven_cat1 = ventas_categorias(tienda)
ven_cat2 = ventas_categorias(tienda2)
ven_cat3 = ventas_categorias(tienda3)
ven_cat4 = ventas_categorias(tienda4)
```

# 3. Calificación promedio de la tienda


```python
def calificacion_promedio(tienda):
  return round(tienda["Calificación"].mean(),1)
cal_prom1 = calificacion_promedio(tienda)
cal_prom2 = calificacion_promedio(tienda2)
cal_prom3 = calificacion_promedio(tienda3)
cal_prom4 = calificacion_promedio(tienda4)

tiendas= ['Tienda1','Tienda2','Tienda3','Tienda4']
calificaciones = [cal_prom1,cal_prom2,cal_prom3,cal_prom4]

plt.plot(tiendas, calificaciones, marker='o', linestyle='--', color='g')
plt.title('Calificación promedio de las tiendas')
plt.xlabel('Tiendas')
plt.ylabel('Calificación promedio')
plt.show()
```

# 4. Productos más y menos vendidos

```python
def productos_mas_menos_vendidos(tienda):
  maximo= tienda.groupby("Producto")["Precio"].sum().sort_values(ascending=False).head(1)
  minimo= tienda.groupby("Producto")["Precio"].sum().sort_values(ascending=False).tail(1)
  return maximo, minimo
tienda1=productos_mas_menos_vendidos(tienda)
tienda2=productos_mas_menos_vendidos(tienda2)
tienda3=productos_mas_menos_vendidos(tienda3)
tienda4=productos_mas_menos_vendidos(tienda4)

maximo, minimo = productos_mas_menos_vendidos(tienda)

# Preparar datos para el gráfico
productos = maximo.index.tolist() + minimo.index.tolist()
valores = maximo.values.tolist() + minimo.values.tolist()

# Crear el gráfico de barras
plt.figure(figsize=(8, 5))
plt.bar(productos, valores, color=['green', 'red'])
plt.title('Productos Más y Menos Vendidos')
plt.xlabel('Productos')
plt.ylabel('Suma de Precios')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
def envio_promedio(tienda):
  return round(tienda["Costo de envío"].mean(),1)

envio_promedio1 = envio_promedio(tienda)
envio_promedio2 = envio_promedio(tienda2)
envio_promedio3 = envio_promedio(tienda3)
envio_promedio4 = envio_promedio(tienda4)

data= {'Tiendas':['Tienda1','Tienda2','Tienda3','Tienda4'],
       'Envios':[envio_promedio1,envio_promedio2,envio_promedio3,envio_promedio4]}

df = pd.DataFrame(data)

# Graficar el envío promedio en un gráfico de barras
plt.figure(figsize=(5, 4))
plt.bar(df['Tiendas'], df['Envios'], color='red')
plt.title('Costo de envío promedio por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Costo de Envío Promedio')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
